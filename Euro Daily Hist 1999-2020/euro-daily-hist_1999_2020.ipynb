```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

- We rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
- We change the Time column to a datetime data type.
- We sort the values by Time in ascending order.
- We reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar'].astype('float')
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.


```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

```python
import datetime as dt
euro_to_dollar = euro_to_dollar[euro_to_dollar['Time'].dt.year.between (2001, 2021)]
```

```python
bush = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2001,2008)]
obama = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2009,2016)]
trump = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2017,2021)]
```

```python
import matplotlib.style as style

# Set the style
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize= (11, 4))

# Plot the line chart
for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.plot(bush['Time'], bush['rolling_mean'], color='purple', linewidth=4.5)
ax.plot(obama['Time'], obama['rolling_mean'],color='orange', linewidth=4.5)
ax.plot(trump['Time'], trump['rolling_mean'], color='blue', linewidth=4.5)

# Modify the x axis
ax.set_xticks(['2001-01-01', '2009-01-01', '2017-01-01', '2021-01-01'])
ax.set_xticklabels(['2001', '2009', '2017', '2021'])

# Modify the y axis
ax.set_yticks([1.0, 1.2, 1.4, 1.6])

# Add title and subtitle
ax.text(10500, 1.77, 'EURO-USD rate averaged 1.22 under the last three US presidents', size = 18, weight='bold')
ax.text(10500,1.7, 'EURO-USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017)', size = 16)
ax.text(10500,1.65, 'and Donald Trump (2017-2021)', size = 16)

# Add president name for each peroid
ax.text(12450,1.5, 'BUSH', size = 16, weight= 'bold', color = 'purple')
ax.text(15300,1.5, 'OBAMA', size = 16, weight= 'bold', color = 'orange')
ax.text(17550,1.5, 'TRUMP', size = 16, weight= 'bold', color = 'blue')

# Add 
ax.text(10500, 0.75,
        '©CHIENNGUYEN' + ' '*133 + 'Source: DATAQUEST',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=12)
plt.show()
```
